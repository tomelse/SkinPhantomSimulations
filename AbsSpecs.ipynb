```python
from spectra import blood_mu_a, melanosome_mu_a, water_mu_a, lipid_mu_a
import matplotlib.pyplot as plt
import numpy as np
import matplotlib

matplotlib.rcParams['figure.figsize'] = (4,4)
matplotlib.rcParams.update({'font.size': 14})
```

```python
wavelengths = np.linspace(500, 1050, num=60)*1e-9
hb = blood_mu_a(wavelengths, 0, water = False)
hbo2 = blood_mu_a(wavelengths, 1, water = False)
water = water_mu_a(wavelengths)
melanin = melanosome_mu_a(wavelengths, 0.1)
lipid = lipid_mu_a(wavelengths)
lipid_mu_a
```

```python
plt.semilogy(wavelengths*1e9, hb, label="Hb")
plt.text(900, 100, "Hb", c="C0")
plt.semilogy(wavelengths*1e9, hbo2, label="HbO2")
plt.text(520, 100, "HbO$_2$", c="C1")
plt.semilogy(wavelengths*1e9, melanin, label="Melanin")
plt.text(700, 5000, "Melanin", c="C2")
plt.semilogy(wavelengths*1e9, water, label="Water")
plt.semilogy(wavelengths*1e9, lipid, label="Lipids")
plt.xlabel("Wavelength / $nm$")
plt.ylabel(r"$\mu_a$ / $m^{-1}$")
plt.show()
```

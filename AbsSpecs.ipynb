```python
from spectra import blood_mu_a, melanosome_mu_a, water_mu_a
import matplotlib.pyplot as plt
import numpy as np
```

```python
wavelengths = np.linspace(500, 1200, num=60)*1e-9
hb = blood_mu_a(wavelengths, 0, water = False)
hbo2 = blood_mu_a(wavelengths, 1, water = False)
water = water_mu_a(wavelengths)
melanin = melanosome_mu_a(wavelengths, 0.1)
```

```python
plt.semilogy(wavelengths*1e9, hb, label="Hb")
plt.semilogy(wavelengths*1e9, hbo2, label="HbO2")
plt.semilogy(wavelengths*1e9, melanin, label="Melanin")
plt.semilogy(wavelengths*1e9, water, label="Water")
plt.xlabel("Wavelength / $nm$")
plt.ylabel("Absorption coefficient / $mm^{-1}$")
plt.legend()
plt.show()
```
